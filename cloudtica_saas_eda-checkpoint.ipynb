```python
# Data Handling
import pandas as pd
import numpy as np

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Statistics
from scipy import stats

# Outlier Detection / Scaling
from sklearn.preprocessing import StandardScaler

# SQL Connection
from sqlalchemy import create_engine

# Display setting
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)

#Style
sns.set(style="whitegrid")

print("Libraries imported successfully")
```

```python
# Load dataset
df = pd.read_csv('Cloudtica.csv')

#Preview data
df.head()
```

```python
# Check dataset shape
df.shape
```

```python
# Check missing
df.info()
```

```python
# Column rename
df = df.rename(columns={'Account Manger':'Account Manager'})
df.columns
```

```python
# Check each column statistic descriptive
display(df.describe(), df.describe(include='object'))
```

```python
# Convert to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'],format='mixed')
df.info()
```

```python
print(df['Order Date'])
```

```python
# Extract useful time features
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
df['Quarter'] = df['Order Date'].dt.quarter
```

```python
# Check missing values
df.isnull().sum()
```

```python
# Drop unnecessary columns
df.drop(columns=['Row ID','License','Date Key'], inplace=True)
df.head()
```

```python
# Check Duplicate records
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
# Statistical Analysis
df[['Sales','Profit','Discount']].agg(['mean','median','std','min','max'])
```

```python
# Outlier detection
columns = ['Sales', 'Profit', 'Discount']
for col in columns:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
# Removing outlier using IQR method
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3-Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return df[(df[column] >= lower) & (df[column] <= upper)]
for col in ['Sales','Profit','Discount']:
    df = remove_outliers(df, col)
sns.boxplot(data=df[['Sales','Profit','Discount']])
                    
```

```python
# Correlation matrix
corr = df[['Sales','Profit','Discount']].corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='viridis')
plt.title("Correlation Matrix")
plt.show()
```

```python
# Sales Distribution
plt.figure(figsize=(8,5))
sns.histplot(df['Sales'], bins=30, kde=True)
plt.title("Sales Distribution")
plt.show()
```

```python
# Profit Distribution
plt.figure(figsize=(8,5))
sns.histplot(df['Profit'], bins=30, kde=True)
plt.title("Profit Distribution")
plt.show()
```

```python
# Sales by Region
region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
region_sales.plot(kind='bar',figsize=(8,5))
plt.title("Sales by Region")
plt.ylabel("Total Sales")
plt.show()
```

```python
# Sales by Industry
industry_sales = df.groupby('Industry')['Sales'].sum().sort_values()
industry_sales.plot(kind='barh',figsize=(8,6))
plt.title("Sales by Industry")
plt.show()
```

```python
# Sales by Product
product_profit = df.groupby('Product')['Profit'].sum().sort_values()
product_profit.plot(kind='barh',figsize=(8,6))
plt.title("Sales by Product")
plt.show()
```

```python
# Monthly Sales Trend
monthly_sales = df.groupby('Month')['Sales'].sum()
plt.figure(figsize=(10,5))
sns.lineplot(x=monthly_sales.index,y=monthly_sales.values)
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
# Top 10 Customers by Sales
top_customers = df.groupby('Customer')['Sales'].sum().sort_values(ascending=False).head(10)
top_customers.plot(kind='bar')
plt.title("Top 10 Customers by Sales")
plt.show()
```
